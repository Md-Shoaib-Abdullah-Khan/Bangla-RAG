```python
from langchain.docstore.document import Document
import pdfplumber
import pytesseract
from PIL import Image

def ocr_bangla_pdf(pdf_path):
    docs = []
    pytesseract.pytesseract.tesseract_cmd = r'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            image = page.to_image(resolution=300).original
            text = pytesseract.image_to_string(image, lang='ben')  # Use 'ben' for Bengali
            if text.strip():
                docs.append(Document(page_content=text.strip(), metadata={"page": page.page_number}))
    return docs

```

```python
import re

def extract_filtered_bangla_mcqs(text):

    questions = re.split(r'\n?\d+[।.]', text)
    mcq_list = []

    for block in questions:
        block = block.strip()
        if not block:
            continue

        block = re.sub(r'\[.*?\]', '', block)

        q_match = re.split(r'[\(\\[]?[কখগঘ][\)\\]]?', block)
        if len(q_match) < 2:
            continue
        question_text = q_match[0].strip()

        options = re.findall(r'[\(\\[]?([কখগঘ])[\)\\]]?\s*([^\n(\\[]+)', block)
        option_dict = {opt[0]: opt[1].strip() for opt in options[:4]}

        answer_match = re.search(r'উত্তর[:ঃ]\s*([কখগঘ])', block)
        answer = answer_match.group(1) if answer_match else None

        filtered_options = {}
        if answer in option_dict:
            filtered_options[answer] = option_dict[answer]

        mcq_list.append(
            f"প্রশ্ন  {question_text} উত্তর: {filtered_options[answer]}",
        )

    return mcq_list
extract_filtered_bangla_mcqs(mcq[5].page_content)
```

```python
final_mcq = " "
for i in mcq:
    final_mcq.join(extract_filtered_bangla_mcqs(i.page_content))
    for j in extract_filtered_bangla_mcqs(i.page_content):
        final_mcq = final_mcq + j.join("\n\n")

print(final_mcq)
```

```python
my_string = final_mcq
file_name = "data/processed.txt"

with open(file_name, 'w', encoding="utf-8") as file:
    file.write(my_string)

print(f"String successfully saved to {file_name}")
```
